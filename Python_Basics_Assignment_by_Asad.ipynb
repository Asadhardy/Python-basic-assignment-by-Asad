```python
# Q7. Describe the different types of loops in python and their use cases with example.
# Answer:
    #In python there are basically 3 types of loops first one is for loop second is while loop and the third is nested loops and are as following:
    
# 1st.For loops:
    # for loop iterates in the particular list, tuple, strings, etc in a sequence, that is define in the program, for loops iterate over every element and return the value present in the list,tuples, string by printing it and for further logic or action in the program
    
# use case of for loop: 
    # for loops print every element one by one in the console, each element will be on a new line.
    
#For loop Example:
animals = ["lion", "tiger", "giraffe", "elephant"]
for animal in animals:
    print(animal)
    
#Insight: This for loop will print all the animals present in the list one by one and, every animal on a new line in the console.

#2nd. While lopp will continue to run when the given condition is True, if the loops become False, the while loop will get terminated(stops and exits)
        #In while loops you have to manually updates the condition otherwise the program will run infinitely, to avoid infinite running of the program the condition, the while loop condition(where to stop or run) has to be set manually by the programmer
        
#while loop use case:
    # if you want to perform an action\operation repeatedly until an certain condition is true,
    # while loop is ideal, which means you dont know how many times the loop will run depending upon your certain condition, and the condition is False the loop will end and exited.
        
#Example:
count = 0
while count <= 10:
    print(count)
    count += 1 #Incrementing counter each time
        
#While loop Insight: In this while loop example, i have initialize a while loop, which starts from 0 and the while loop will print until the count value is lesser than 10 or equal to 10, it will print count until the while loop becomes False.
        
#3rd. nested loop:
    
#When you have to run a loop inside of another loop is known as nested loop, the outer loop will execute first, in every iteration of the first loop, the inner loop will completely run
        
#Example:
for i in range(2): #outer loop with 2 rows
    for j in range(2): #inner loop with 2 column
        print(f"Row {i}, column {j}")
#Nested loop Insight:When the outer loop will iterate, it will be (i = 0), then inner loop will start iterate, it will be (j = 0 and j = 1) and will print the output:
        #output will be Row 0, column 0
        #               Row 0, column 1
        
        #and when the outer loop will iterate for the second time (given in range(2)) it will (i = 1) , after outer layer, it will iterate nested loop for the second time (j = 1) and print the value
        #Output: Row 1, column 0
        #        Row 1, column 1
        #     it will print the grid columns and rows 2 by 2, which print rows and colums seperately.
```
